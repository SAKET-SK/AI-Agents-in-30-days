```python
import os
import json
from openai import OpenAI
from IPython.display import Markdown, display

client = OpenAI()
```

```python
# The basic story idea - you can change this!
story_idea = "A beautiful day in a busy city, where people are having fun but little they know a catastrophic disaster is about to happen."

print(f"Story Idea: {story_idea}")
```

```python
styles = {
    "Scary": "Write a creepy, scary story. Make it spooky and unsettling. Use dark imagery.",
    "Funny": "Write a funny, humorous story. Make it silly and entertaining. Include jokes or funny situations.",
    "Mysterious": "Write a mysterious, suspenseful story. Make it intriguing like a detective novel. Build curiosity."
}

print("We'll generate these 3 versions:")
for style_name in styles.keys():
    print(f"  - {style_name}")
```

```python
story_versions = []
story_text = []

# Generate each story one at a time
for style_name, style_instructions in styles.items():
    
    print(f"\n{'='*60}")
    print(f"Generating: {style_name} version")
    print(f"{'='*60}\n")
    
    # Create the prompt for GPT
    prompt = f"""{style_instructions}

Write a SHORT story (about 300 words) based on this idea:
{story_idea}

Make it engaging with a beginning, middle, and end."""
    
    # Call the OpenAI API
    response = client.chat.completions.create(
        model="gpt-4o-mini",  # Using the mini model - it's fast and cheap!
        messages=[
            {"role": "user", "content": prompt}
        ],
        temperature=0.8  # Higher = more creative, Lower = more predictable
    )
    
    # Get the story from the response
    story = response.choices[0].message.content
    
    # Show the story
    print(f"**{style_name} Story:**\n")
    display(Markdown(story))
    
    # Save it for later
    story_versions.append(style_name)
    story_text.append(story)
    
    print("\n")
```

```python
# Verify we have 3 stories
print(f"‚úÖ We generated {len(story_versions)} stories:")
for i, name in enumerate(story_versions, 1):
    print(f"   {i}. {name}")
```

```python
# Combine all stories into one big text
all_stories = ""

for number, story in enumerate(story_text, 1):
    all_stories += f"Story #{number}\n\n"
    all_stories += story
    all_stories += "\n\n" + "---" * 20 + "\n\n"

print("Combined all stories! Here's a preview:")
print(all_stories[:300] + "...")
```

```python
# Create instructions for the judge
judge_instructions = f"""You are a judge in a creative writing contest.

You need to read 3 short stories and rank them from best to worst.

Consider these things:
- How engaging is the story?
- How well-written is it?
- Does it match its intended style?
- Is it creative and original?

Here are the 3 stories:

{all_stories}

Now rank them from best to worst.

Respond with ONLY this JSON format (no extra text):
{{
  "rankings": ["1", "2", "3"],
  "why_winner_won": "Explain in 1-2 sentences why the best story won"
}}

The rankings should be story numbers from best to worst."""

print("Judge instructions created!")
```

```python
# Ask the judge to evaluate
print("Asking the AI judge to pick a winner...\n")

judge_response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": judge_instructions}
    ],
    temperature=0.3  # Lower temperature = more consistent judging
)

# Get the judge's decision
judgment = judge_response.choices[0].message.content
print("Judge's response:")
print(judgment)
```

```python
# Parse the JSON response
try:
    # Convert the text response into a Python dictionary
    results = json.loads(judgment)
    
    # Get the rankings and explanation
    rankings = results["rankings"]
    explanation = results["why_winner_won"]
    
    # Display the results!
    print("\n" + "="*60)
    print("üèÜ FINAL RESULTS üèÜ")
    print("="*60 + "\n")
    
    # Show each ranking
    medals = ["ü•á 1st Place", "ü•à 2nd Place", "ü•â 3rd Place"]
    
    for position, story_number in enumerate(rankings):
        story_index = int(story_number) - 1  # Convert to array index
        style_name = story_versions[story_index]
        print(f"{medals[position]}: {style_name} Story")
    
    # Show why the winner won
    winner_index = int(rankings[0]) - 1
    winner_name = story_versions[winner_index]
    
    print("\n" + "="*60)
    print(f"üéâ WINNER: {winner_name} Story! üéâ")
    print("="*60)
    print(f"\nWhy it won: {explanation}\n")
    
except json.JSONDecodeError:
    print("Oops! Couldn't parse the judge's response.")
    print("Raw response:", judgment)
```

```python
# Show the winning story
winner_index = int(rankings[0]) - 1
winner_name = story_versions[winner_index]
winner_story = story_text[winner_index]

print("="*60)
print(f"THE WINNING STORY: {winner_name}")
print("="*60 + "\n")
display(Markdown(winner_story))
```

## What Did We Learn?

**Key Concepts:**

1. **Different Prompts = Different Outputs**
   - The same story idea created 3 totally different stories
   - The prompt/instructions control the style and tone

2. **AI as a Judge**
   - We used AI to evaluate AI-generated content
   - This is called an "evaluator pattern"

3. **Multi-Agent Pattern**
   - We had multiple "agents" (the 3 story writers) work on the same task
   - Then one "judge agent" evaluated them all

4. **Structured Output (JSON)**
   - We asked for JSON so we could easily use the results in code
   - This makes the AI's output more useful for programs

**This pattern is used in real applications for:**
- Generating multiple options and picking the best
- A/B testing different approaches
- Quality control and evaluation
- Creative brainstorming
